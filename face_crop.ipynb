```python
#codingï¼šutf-8
import  cv2
import os
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
```

```python
def face_dect(file_path):
    image = cv2.imread(file_path,1);
    #if len(image.shape) > 2 and image.shape[2] == 3:
        #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    classifier=cv2.CascadeClassifier("/home/qsc/DEMO/VGGFace_TF/haarcascade_files/haarcascade_frontalface_default.xml")
    faces = classifier.detectMultiScale(image)
    if not len(faces) > 0:
        return None
    max_face = faces[0]
    for face in faces:
        if face[2] * face[3] > max_face[2] * max_face[3]:
            max_face = face
    face_image = image[max_face[1]:(max_face[1] + max_face[2]), max_face[0]:(max_face[0] + max_face[3])]
    try:
        image = cv2.resize(face_image, (48, 48), interpolation=cv2.INTER_CUBIC) / 255.
    except Exception:
        print("[+} Problem during resize")
        return None
    return face_image

def load_file(examples_list_file):
    lines = np.genfromtxt(examples_list_file, delimiter="", dtype=[('col1', 'S120'), ('col2', 'i8')])
    #f=open(train_file)
    #lines=f.readlines()
    examples = []
    labels = []
    for example, label in lines:
        examples.append(example)
        labels.append(label)
    return np.asarray(examples), np.asarray(labels)
```

```python
file_path="/home/qsc/DEMO/VGGFace_TF/k-fold/path_label_selected.txt"
imgs_save_path="/home/qsc/DEMO/CASME2_face_crop"
imgs_src_path="/home/qsc/DEMO/CASME2_RAW_selected"

lines = np.genfromtxt(file_path, delimiter="", dtype=[('col1', 'S120'), ('col2', 'i8')])
for example, _ in lines:
    #print(example)
    img_path=str(example)[2:-1]
    print(img_path)
    img=face_dect(img_path)
    save_path=img_path.replace("CASME2_RAW_selected","CASME2_face_crop")
    img_dir,img_name=os.path.split(save_path)
    img_save_name=img_name.strip("'")
    os.makedirs(img_dir,exist_ok=True)
    #print(img_dir+"***"+img_save_name)
    cv2.imwrite(img_dir+'/'+img_save_name,img)
    
"""
examples,_=load_file(file_path)
for i in range(len(examples)):
    img=face_dect(str(example[i]))
    save_name=example[i].replace("CASME2_RAW_selected","CASME2_face_crop")
    print(save_name)
    cv2.imwrite(save_name,img)
    
"""

```
